# Some initial data cleaning

The date columns is just being read as an "object".  Much better to have a datetime object.

```python
crisisData.date = crisisData.date.apply(lambda x: pd.to_datetime(x, infer_datetime_format=True))
crisisData.head()
```

# Initial exploration

Let's just plot some stuff to warm ourselves up to the data.

```python
# Print all the columns within a range
start, end = 0, 50
for i in range(start, end): print(crisisData.columns[i])
```

```python
plt.hist(crisisData.Q100r1)
```

```python
# Looks bad--apparently 97 is used to code non-response.  Let's just filter it to get started.
q100r1 = crisisData.Q100r1[crisisData.Q100r1 < 11]
plt.hist(q100r1)
```

```python
# That looks fine, gives an impression of the data.  
# But just to clarify the graph a little, could be nice to force 
# the bins to be a little better marked.
fig, ax = plt.subplots()
ax.hist(x=q100r1, bins=np.arange(.5,11), rwidth=.9)
ax.set_xticks(ticks=range(1,11))
```

```python
# Let's try slicing out a certain date range. Let's start by seeing what the current range of dates is.
print(crisisData.date.min())
print(crisisData.date.max())
```

```python
# First make a date for the upper bound, selecting only data from the 4th.
fourth = dt.datetime(2020,6,4)
print(fourth)
OneThroughTen = crisisData[crisisData.date <= fourth]

print(OneThroughTen.head())
```

```python
# Oh, it must be that when you create a date, its hour-mark starts at the earliest time during that day.  
# Let's set a later date.
fifth = dt.datetime(2020,6,5)
OneThroughTen = crisisData[crisisData.date < fifth]
```

```python
print(OneThroughTen.date.min())
print(OneThroughTen.date.max())
```

```python
# Let's see how that question looks differently on the fourth.  
q100r1 = OneThroughTen.Q100r1[OneThroughTen.Q100r1 <= 10]
plt.hist(q100r1)
```
